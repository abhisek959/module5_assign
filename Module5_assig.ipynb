Q6. Explain the concept of generators in Python and how they are defined.

A-. Generators in Python are a special type of iterator that allow you to iterate over a sequence of values lazily (one at a time, only when needed), which makes them memory-efficient and ideal for handling large datasets or infinite sequences.

A generator is a function that:

Uses the yield keyword instead of return

Automatically creates an iterator

Pauses its state between values and resumes when next() is called.

Using a Generator Function:
Defined like a regular function, but uses yield.

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

Q9.  Explain the purpose and usage of the `map()` function in Python.

A-> The map() function in Python is a built-in higher-order function used to apply a given function to all items in an iterable (like a list or tuple), returning a new iterator with the results.

Purpose of map()
To transform elements of an iterable by applying a function to each element.

It’s an efficient and concise way to perform operations on collections without writing explicit loops.



Q.10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

```python
"""Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list."""

def sum_of_even_numbers(numbers):
    even_sum = sum(num for num in numbers if num % 2 == 0)
    return even_sum

nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_even_numbers(nums))  # Output: 12
```

```python
# Create a Python function that accepts a string and returns the reverse of that string

def reverse_string(text):
    return text[::-1]

my_str = "Hello, World!"
reversed_str = reverse_string(my_str)
print("Reversed string:", reversed_str)
```

```python
""" Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number."""

def square_numbers(numbers):
    return [num ** 2 for num in numbers]

nums = [1, 2, 3, 4, 5]
squares = square_numbers(nums)
print("Squared numbers:", squares)
```

```python
"""Write a Python function that checks if a given number is prime or not from 1 to 200"""

def is_prime(n):
    if n < 2 or n > 200:
        return False  # Not in the valid range
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(7))
print(is_prime(100))
print(is_prime(199))


```

```python
"""Write a generator function in Python that yields the powers of 2 up to a given exponent."""

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Generate powers of 2 up to 5 (2^0 to 2^5)
for power in powers_of_two(5):
    print(power)

```

```python
""" Implement a generator function that reads a file line by line and yields each line as a string"""

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


for line in read_file_line_by_line(file_path):
    print(line)

```

```python
""" Use a lambda function in Python to sort a list of tuples based on the second element of each tuple."""

# List of tuples
data = [(1, 4), (3, 1), (5, 2), (2, 3)]

# Sort based on the second element of each tuple
sorted_data = sorted(data, key=lambda x: x[1])

print("Sorted list:", sorted_data)
```

```python
"""Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit."""

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 10, 20, 30, 40]

# Use map() to convert to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

```python
"""Create a Python program that uses `filter()` to remove all the vowels from a given string."""

def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    # Use filter to keep only non-vowel characters
    result = ''.join(filter(lambda char: char not in vowels, input_str))
    return result

text = "Hello, World!"
no_vowels = remove_vowels(text)
print("String without vowels:", no_vowels)
```
